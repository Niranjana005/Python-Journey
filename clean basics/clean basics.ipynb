```python
#printing the largest number among three
a=int(input())
b=int(input())
c=int(input())
if a>b and a>c:
    print(f'{a} is the largest number')
elif b>a and b>c:
    print(f'{b} is the largest number')
elif a==b==c:
    print("equal values") 
else:
    print(f'{c} is the largest number')
```

```python
#a simple calculator operations with 5 operations
a=int(input())
b=int(input())
print("enter any operation on +,-,*,/,^:")
c=input()
if c=="+":
    print("Addition:",a+b)
elif c=="-":
    print("Subtraction:",a-b)
elif c=="*":
    print("Multiplication:",a*b)
elif c=="/":
    print("Division:",a/b)
elif c=="^":
    print(f"Power of {a}:",a**b)
else:
    print("invalid operation")
```

```python
#grade calculation
a=int(input())
if 90<a<=100:
    print("O Grade")
elif 80<=a<=90:
    print("A Grade")
elif 70<=a<=80:
    print("B Grade")
elif 60<=a<=70:
    print("C Grade")
elif 50<=a<=60:
    print("D Grade")
elif a<50:
    print("Fail")
else:
    print("Enter a valid mark out of 100")
```

```python
#leap year check
a=int(input())
if a%400==0:
    print("its a leap year")
elif a%4==0 and a%100!=0:
    print("its a leap year")
else:
    print("its not a leap year")
```

```python
#palindrome check
a=input()
if a==a[::-1]:
    print(f'{a} is a palindrome')
else:
    print(f'{a} is not a palindrome')
```

```python
#counting the numer of digits in the number
a=int(input())
b=str(a)
count=0
for i in b:
    if i.isdigit()==True:
        count+=1
print(count)
```

```python
#printing common numbers in the three lists
a=list(map(int,input().split()))
b=list(map(int,input().split()))
c=list(map(int,input().split()))
for i in a:
    for j in b:
        for k in c:
            if i==j==k:
                print(i,end=" ")
```

```python
#splitting the odd and even numbers from an array
a=int(input())
arr=[]
for i in range(a):
    arr.append(int(input()))
    odd=[]
    even=[]
    for j in arr:
        if j%2==0:
            even.append(j)
        else:
            odd.append(j)
print("The odd elements are",odd)
print("The even elements are",even)
```

```python
#sum of the digits in the number
a=(input())
summ=[]
for i in a:
    summ.append(int(i))
tot=0
for j in summ:
    tot+=j
print("The sum of the digits in a number is:",tot)
    
```

```python
# min and max number in a list
a=list(map(int,input().split()))
print("The maximum thing in the list is:",max(a))
print("The minimum thing in the list is:",min(a))
```

```python
#finding second largest number in the list
a=list(map(int,input().split()))
max_=max(a)
res=[]
for i in a:
    if i<max_:
        res.append(i)
print("The second largest number in the list is:",max(res))
```
