```python
#basic triangular star pattern
a=int(input())
for i in range(a):
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
#inverted star triangle
a=int(input())
for i in range(a):
    for j in range(a-i):
        print("*",end=" ")
    print()
```

```python
#right sided upward triangle
a=int(input())
for i in range(a):
    for j in range(a-i-1):
        print(" ",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
#right sided inverted triangle
a=int(input())
for i in range(a-1,-1,-1):
    for j in range(a-i):
        print(" ",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
#alphabetical upward triangle
a=int(input())
ch=ord("A")
for i in range(a):
    for j in range(i+1):
        print(chr(ch),end=" ")
    print()
```

```python
#alphabetical upward triangle with column increment
a=int(input())
ch=ord("A")
for i in range(a):
    for j in range(i+1):
        print(chr(ch+j),end=" ")
    print()
    
```
